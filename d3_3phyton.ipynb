```python
import matplotlib.pyplot as plt
import numpy as np

def plot_colormaps():
    colormaps = plt.colormaps()
    n = len(colormaps)
    fig, axs = plt.subplots(n//2, 2, figsize=(10, 2*(n//2)))
    for ax, cmap_name in zip(axs.ravel(), colormaps):
        cmap = plt.get_cmap(cmap_name)
        img = np.linspace(0, 1, 256).reshape(1, -1)
        ax.imshow(img, aspect='auto', cmap=cmap)
        ax.set_title(cmap_name, fontsize=8)
        ax.set_axis_off()
    plt.show()

plot_colormaps()
```

```python
import requests
from bs4 import BeautifulSoup #internet sitelerinden bilgi çeker

url = 'https://barislx.blogspot.com/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        text = link.text.strip()
        if href:
            print("Bağlantı metni: ", text, "URL:", href)
    print("Toplam bulunan bilgi sayısı: ", len(links), " adet")
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
image = cv2.imread('Opencv_ornekleri/Siberakademi.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görüntünün Gösterimi")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (9, 9), 0)

plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python

edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama")

plt.show()
```
